```python
import dotenv
import os
from sqlalchemy import create_engine
import pandas as pd

"QUESTe SONO LE CREDEDENZIALI CONTENUTE NEL FILE .ENV"
dotenv.load_dotenv() 
#dotenv.load_dotenv(override=True) #sovrascrive le variabili di sistema con quelle di ambiente tipo lo username

username=os.getenv("usernames")
password=os.getenv("password")
host=os.getenv("host")
dbname=os.getenv("dbname")


conn_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"
print(conn_string)

db_engine = create_engine(conn_string)

query = """ SELECT *
            FROM dimproduct
        """
df = pd.read_sql(query, db_engine)
db_engine.dispose() #istruzione per chiudere il database
df
```

```python
#Esercizio: Il dataset diabetes.csv raccoglie persone con diabete o meno, 
#e il valore di diverse variabili fisiologiche dei pazienti.
df_diabetes = pd.read_csv("diabetes.csv")

#Osserviamone le dimensioni e un'anteprima di cinque righe;
df_diabetes.shape
```

```python
#Prendiamoci un po' di tempo per dare un'occhiata ai metadati delle colonne;
df_diabetes.dtypes
```

```python
#Selezioniamo i dati relativi a diverse fasce di età: <20, 20-30, 30-40, 40-50, >50;
#Qual è la media della pressione sanguigna diastolica per le diverse fasce di età? 

ranges = [0, 20, 30, 40, 50, 100]

data_collection = []

for index, num in enumerate(ranges[:-1]):
    
    data = df_diabetes[(df_diabetes["Age (years)"] > num) & (df_diabetes["Age (years)"] <= ranges[index + 1])][ "Diastolic blood pressure (mm Hg)"]
    data_collection.append(data)
    mean_for_range = data.mean()
    
    if num == 50:
        print(f"Media pressione diastolica in soggetti > 50 anni: {mean_for_range:.2f}")
        break
    
    print(f"Media pressione diastolica in fasce età tra {num} e {ranges[index + 1]}: {mean_for_range:.2f}\n")
```

```python
#Qual è la media della pressione per ogni singolo anno di età?
start = df_diabetes["Age (years)"].min()
stop = df_diabetes["Age (years)"].max() + 1

for age in range(start, stop):
    
    pressure_mean = df_diabetes[df_diabetes['Age (years)'] == age][ 'Diastolic blood pressure (mm Hg)'].mean()
    print(f"Età {age}, Pressione Diastolica {pressure_mean:.2f}")
```
